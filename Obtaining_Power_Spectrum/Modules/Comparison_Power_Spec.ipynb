```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
cdm_path = '/home/isha/Desktop/AU/Work_Done/Wave_Dark_Matter/Obtaining_Power_Spectrum/CDM/'
fdm_path = '/home/isha/Desktop/AU/Work_Done/Wave_Dark_Matter/Obtaining_Power_Spectrum/FDM/'
fdm_si_path = '/home/isha/Desktop/AU/Work_Done/Wave_Dark_Matter/Obtaining_Power_Spectrum/FDM+SI/'
output_path = '/home/isha/Desktop/AU/Work_Done/Wave_Dark_Matter/Obtaining_Power_Spectrum/'
```

```python
cdm = pd.read_csv(cdm_path + 'cdm_density_growth_allkvals.csv')
fdm = pd.read_csv(fdm_path + 'fdm_density_growth_allkvals.csv')
fdm_si = pd.read_csv(fdm_si_path + 'fdm_si_density_growth_allkvals.csv')
```

```python
cdm = pd.DataFrame(cdm)
fdm = pd.DataFrame(fdm)
fdm_si = pd.DataFrame(fdm_si)


```

```python
cdm['P_k'] = cdm['delta_cdm'] **2
fdm['P_k'] = fdm['delta_fdm'] **2
fdm_si['P_k'] = fdm_si['delta_fdm'] **2
```

```python
cdm =  cdm[cdm['t'] == cdm['t'].max()]
fdm =  fdm[fdm['t'] == fdm['t'].max()]
fdm_si =  fdm_si[fdm_si['t'] == fdm_si['t'].max()]

cdm, fdm, fdm_si
```

```python
plt.plot(cdm['k'], cdm['P_k'], label='CDM')
plt.plot(fdm['k'], fdm['P_k'], label='FDM')
plt.plot(fdm_si['k'], fdm_si['P_k'], label='FDM+SI')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('k (h/Mpc)')
plt.ylabel('P(k) ((Mpc/h)^3)')
plt.title('Power Spectrum Comparison at Final Time Step')
plt.legend()
plt.grid(True)
#plt.savefig(output_path + 'Power_Spectrum_Comparison.png', dpi=300)
plt.show()  
```
